## Mergeo Datos en un solo frame para usarlos mas facil

```python
data = []
files=[
    "1994.csv",
    "1995.csv",
    "1996.csv",
    "1997.csv",
    "1998.csv",
    "1999.csv",
    "2000.csv",
    "2001.csv",
    "2002.csv",
    "2003.csv",
    "2004.csv",
    "2005.csv",
    "2006.csv",
    "2007.csv",
    "2008.csv" 
]
root="../data/"
for file in files:
    try:
        df = pd.read_csv(root+file)
        data.append(df)
    except:
        print('error con '+file)
frame = pd.concat(data, axis=0, ignore_index=True)
```
